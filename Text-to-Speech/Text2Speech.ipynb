```python
from pathlib import Path
from openai import OpenAI
import os
from os import environ
import time

client = OpenAI()
client.api_key = os.getenv("OPENAI_API_KEY")

# Define the path to save the speech file using the current working directory
speech_file_path = Path(os.getcwd()) / "speech.mp3"

def create_speech():
    try:
        response = client.audio.speech.create(
            model="tts-1",
            voice="nova",
            input="Out of the night that covers me, Black as the pit from pole to pole, I thank whatever gods may be, For my unconquerable soul. In the fell clutch of circumstance, I have not winced nor cried aloud. Under the bludgeonings of chance, My head is bloody, but unbowed. Beyond this place of wrath and tears, Looms but the Horror of the shade, And yet the menace of the years, Finds and shall find me unafraid. It matters not how strait the gate, How charged with punishments the scroll, I am the master of my fate, I am the captain of my soul."
        )
        response.stream_to_file(speech_file_path)
        print("Speech file created successfully.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Retry mechanism
max_retries = 3
for attempt in range(max_retries):
    try:
        create_speech()
        break  # Exit loop if successful
    except Exception as e:
        print(f"Attempt {attempt + 1} failed: {e}")
        time.sleep(5)  # Wait for 5 seconds before retrying
        if attempt == max_retries - 1:
            print("Max retries reached. Please check your connection and API status.")
```
